```python
# Create a subset

df = ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
# define price_label function to return low- mid- and high-range products by price

def price_label(row):

    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range'
    else: return 'Not enough data'
```

```python
# Add new column 'price_range' to df to store results of price_label()

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check the value count in the new column (count of 'Mid-range, Low-range, High-range products)')

df['price_range'].value_counts(dropna = False)
```

```python
# Check the max price in 'prices' column (should be < 15)
df['prices'].max()
```

# If-Statements with the loc() Function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

# If-Statements with For-Loops

Create a new column in your ords_prods_merge dataframe that summarizes how busy each day of the week is

To start, you need to know on which day most orders take place. You can find this out by printing the frequency of the “orders_day_of_the_week” column, which would look like this:

```python
# print the frequency of 'orders_day_of_the_week'

ords_prods_merged['orders_day_of_the_week'].value_counts(dropna = False)
```

```python
# 0 = Saturday
# 1 = Sunday
# 2 = Monday
# 3 = Tuesday
# 4 = Wednesday
# 5 = Thursday
# 6 = Friday

# create an empty list, result
result = []

# for loop

for value in ords_prods_merged["orders_day_of_the_week"]:
    if value == 0:
        result.append("Busiest day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")
    
```

```python
result
```

```python
# Create a new column called 'busiest_day' in 'ords_prods_merged' for result

ords_prods_merged['busiest_day'] = result
```

```python
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

# Question 4
*When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”*

```python
# Check frequency of 'order_hour_of_day' column
# Midnight = 0

ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

**10am** is the **busiest hour**
</br>
**3am** is the **least busy hour** *etc.*

```python
# Periods of time:
# "Most orders" = top 6 hours are 10:00 – 15:00 (6 hours total)
# "Fewest orders" = bottom 6 hours are 00:00 – 5:00 (6 hours total) 
# "Average orders" = others 16:00 – 23:00 (12 hours total)

result = []

for value in ords_prods_merged["order_hour_of_day"]:
    if value == 10:
        result.append("Most orders")
    elif value == 11:
        result.append("Most orders")
    elif value == 12:
        result.append("Most orders")
    elif value == 13:
        result.append("Most orders")
    elif value == 14:
        result.append("Most orders")
    elif value == 15:
        result.append("Most orders")
  
    elif value <= 5:
        result.append("Fewest orders")
    else:
        result.append("Average orders")
```

```python
# Check result
result
```

```python
# Create a new column called 'busiest_days' (plural) in 'ords_prods_merged' for result

ords_prods_merged['busiest_period_of_day'] = result
```
